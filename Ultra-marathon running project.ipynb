```python
df2 = df[
    (df['Event distance/length'].isin(['50km', '50mi'])) &
    (df['Year of event'] == 2020) &
    (df['Event name'].str.extract(r'\((.*?)\)')[0] == 'USA')
]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
#Removing USA from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
#Clean up athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
#Droping colums like Athlete Club, Athlete Country, Athlete year of birth and Athlete Age Category
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth','Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
#Clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
#Check for duplicate values
```

```python
df2[df2.duplicated() == True]
```

```python
#Reset index
```

```python
df2.reset_index(drop = True)
```

```python
#Fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
#Rename columns
```

```python
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates':'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performance',
                            'Athlete gender':'athlete_gender', 
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id'
                           })
```

```python
df2.head()
```

```python
#Reorder columns
```

```python
df3 = df2[['race_day', 'race_name', 'race_length','race_number_of_finishers', 'athlete_id','athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

```python
df3.head()
```

```python
#Charts and graphs
```

```python
#Questions I want to find out from the data
```

```python
#Difference in speed for the 50k,50 mi between Male and Female
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
#What age groups are the best in the 50 km race(20 + races min)
```

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=False).query('count > 19').head(15)
```

```python
#What age groups are the worst in the 50 km race(10 + races min)
```

```python
df3.query('race_length == "50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending=True).query('count > 19').head(15)
```

```python
#Seasons for the data, Slower in summer than winter?
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] =  df3['race_month'].apply(lambda x: 'Winter' if x > 11 else ' Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```

```python
#50 km only
```

```python
df3.query('race_length == "50km"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```
